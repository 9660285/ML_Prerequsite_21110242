Question 1
Given matrices of dimensions (n, k) and (k, m), matrix multiplication involves multiplying elements along the rows of the first matrix and the columns of the second matrix, followed by summation. For each element in the resulting matrix of dimensions (n, m), the total multiplications required are k, and the total additions needed are k-1. Consequently, with a final matrix of dimensions (n, m), the overall number of multiplications is n * m * k, and the total additions amount to n * m * (k-1).

```python
#Question 2a

#let us assume a 3x3 and 3x3 matrix for simplicity. Solved using nested lists
import time

A = [[1,2,3],[4,5,6],[7,8,9]]
B = [[7,8,9],[4,5,6],[1,2,3]]

start = time.time_ns()

rows_A = len(A)
cols_A = len(A[0])
rows_B = len(B)
cols_B = len(B[0])

ans = [[0 for i in range(cols_B)] for i in range(rows_A)]

for i in range(rows_A):
    for j in range(cols_B):
        for k in range(cols_A):
            ans[i][j] += A[i][k] * B[k][j]

end = time.time_ns()

print("Time Taken:" ,end - start,"ns")

for i in ans:
    print(i)
```

```python
#Question 2b

#Solved using numpy

import numpy as np
import time

A = [[1,2,3],[4,5,6],[7,8,9]]
B = [[7,8,9],[4,5,6],[1,2,3]]

start = time.time_ns()

rows_A = len(A)
cols_A = len(A[0])
rows_B = len(B)
cols_B = len(B[0])

ans = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

ans = np.dot(A,B)

end = time.time_ns()

print("Time Taken:" ,end - start,"ns")

for i in ans:
    print(i)
```

It is evident that the numpy code executes more quickly than the nested list code. This is due to the fact that C, which is faster than Python and more effective and optimised for numerical operations, is used to implement numpy. The difference in the time required to multiply the matrices increases as they get bigger and bigger.

```python
Question 4
Gradient w.r.t. x:
2*x*y + (y^3)*cos(x)

Gradient w.r.t. y:
x^2 + 3*(y^2)*sin(x)
```

```python
#Question 5
import jax
import numpy as np
import math

def my_function(x, y):
    return (x**2) * y + (y**3)*jax.numpy.sin(x)


def analytical_solution(x, y):
    grad_x = 2 * x * y + (y**3) * math.cos(x)
    grad_y = x**2 + 3 * (y**2) * math.sin(x)
    return grad_x, grad_y


def gradient_function(x_val, y_val):
    grad_f_x = jax.grad(my_function, argnums=0)
    grad_f_y = jax.grad(my_function, argnums=1)  

    gradient_at_x = grad_f_x(x_val, y_val)
    gradient_at_y = grad_f_y(x_val, y_val)
    
    return float(gradient_at_x), float(gradient_at_y)

x_value = 2.0
y_value = 3.0

analytical_gradients = analytical_solution(x_value, y_value)

print(analytical_gradients)
print(gradient_function(x_value,y_value))
```

```python
#Question 8
class Student:
    def __init__(self, roll_number, name):
        self.roll_number = roll_number
        self.name = name
        self.marks = {}

    def add_subject_mark(self, subject, mark):
        self.marks[subject] = mark

class Branch:
    def __init__(self):
        self.students = []

    def add_student(self, student):
        self.students.append(student)

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = {}

    def add_branch(self, branch_name):
        self.branches[branch_name] = Branch()

    def add_student_to_branch(self, branch_name, student):
        if branch_name in self.branches:
            self.branches[branch_name].add_student(student)

year_2022 = Year(2022)
year_2022.add_branch('Branch 1')
year_2022.add_branch('Branch 2')

student1 = Student(42, 'Yash')
student1.add_subject_mark('Maths', 100)
student1.add_subject_mark('English', 70)
student1.add_subject_mark('Physics', 90)
student1.add_subject_mark('Chemistry', 82)
year_2022.add_student_to_branch('Branch 1', student1)

student2 = Student(34, 'Naman')
student2.add_subject_mark('Maths', 90)
student2.add_subject_mark('English', 95)
student2.add_subject_mark('Physics', 75)
student2.add_subject_mark('Chemistry', 88)
year_2022.add_student_to_branch('Branch 2', student2)

year_2023 = Year(2023)
year_2023.add_branch('Branch 1')
year_2023.add_branch('Branch 2')

student3 = Student(1, 'Aryan')
student3.add_subject_mark('Maths', 100)
student3.add_subject_mark('English', 70)
year_2023.add_student_to_branch('Branch 1', student3)

student4 = Student(31, 'Ketan')
student4.add_subject_mark('Maths', 90)
student4.add_subject_mark('English', 95)
year_2023.add_student_to_branch('Branch 2', student4)

print(year_2022.branches['Branch 1'].students[0].name)
print(year_2023.branches['Branch 2'].students[0].marks)
```

Broadcasting is a feature in numpy that allows for arithmetic operations between arrays of different shapes and sizes. 

```python
#Question 10
import numpy as np
import pandas as pd

# A matrix of size 20x5 with random numbers in the range [1, 2)
random_matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(random_matrix, columns=["a", "b", "c", "d", "e"])

print("DataFrame:")
print(df)
 
# The column with the highest standard deviation
max_std_column = df.std().idxmax()
print("\nColumn with the highest standard deviation:", max_std_column)

# The row with the lowest mean
min_mean_row = df.mean(axis=1).idxmin()
print("Row with the lowest mean:", min_mean_row)
```

```python
#Question 11
import numpy as np
import pandas as pd

# A matrix of size 20x5 with random numbers in the range [1, 2)
random_matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(random_matrix, columns=["a", "b", "c", "d", "e"])

# Add a new column "f" which is the sum of columns "a", "b", "c", "d", "e"
df["f"] = df["a"] + df["b"] + df["c"] + df["d"] + df["e"]

# Add a new column "g" with values "LT8" or "GT8" based on the value in column "f"
df["g"] = np.where(df["f"] < 8, "LT8", "GT8")

print("Updated DataFrame:")
print(df)

# Number of rows where the value in column "g" is "LT8"
count_LT8 = df[df["g"] == "LT8"].shape[0]
print("\nNumber of rows where 'g' is 'LT8':", count_LT8)

# The standard deviation of column "f" for rows where "g" is "LT8"
std_dev_LT8 = df[df["g"] == "LT8"]["f"].std()
print("Standard deviation of 'f' for 'LT8' rows:", std_dev_LT8)

# The standard deviation of column "f" for rows where "g" is "GT8"
std_dev_GT8 = df[df["g"] == "GT8"]["f"].std()
print("Standard deviation of 'f' for 'GT8' rows:", std_dev_GT8)
```

```python
#Quesiton 13
import numpy as np

def custom_argmin(arr):
    min_index = 0

    for i in range(1, len(arr)):
        if arr[i] < arr[min_index]:
            min_index = i

    return min_index

array = np.array([5,32,7,2,6,9,10])

index = custom_argmin(array)
print("argmin result:", index)

# Using np.argmin for verification
index_np = np.argmin(array)
print("np.argmin result:", index_np)
```
